```python
import pandas as pd
import decimal
import numpy as np
```

```python
# Study data files
ev_data_df = pd.read_csv("DATA/Electric_Vehicle_Population_Data.csv")
ev_data_df.head()
```

```python
# Split the 'Vehicle Location' column into 'Lat' and 'Lon' columns using regular expression
ev_data_df['Lon'] = ev_data_df['Vehicle Location'].str.extract('(-?\d+\.\d+)')[0].astype(float)
ev_data_df['Lat'] = ev_data_df['Vehicle Location'].str.extract('(\d+\.\d+)\)')[0].astype(float)
ev_data_df.head()
```

```python
ev_data_df.info()
```

```python
# # Find the duplicate rows by VIN (1-10)

duplicates = ev_data_df[ev_data_df.duplicated(subset=['VIN (1-10)'])]

# display duplicate row
duplicates

```

```python
# Count the number of duplicate rows for each 'VIN (1-10)'
counts = duplicates['VIN (1-10)'].value_counts()
counts
```

```python
# Create a clean DataFrame by dropping the duplicate VIN (1-10) 
ev_data_df.drop_duplicates(subset=['VIN (1-10)'], inplace=True)
```

```python
# Change the data type in to string for City,State,Make
ev_data_df[["County","City","State","Make","Electric Vehicle Type"]]=ev_data_df[["County","City","State","Make","Electric Vehicle Type"]].astype(str)
```

```python
# check the data types for the columns
print (type(ev_data_df['Make'].iloc[0]))
print (type(ev_data_df['State'].iloc[0]))
print (type(ev_data_df['City'].iloc[0]))
print (type(ev_data_df['Electric Vehicle Type'].iloc[0]))
print (type(ev_data_df['County'].iloc[0]))
```

```python
clean_ev_data_df = ev_data_df.drop(["Base MSRP","Legislative District"], axis=1)
clean_ev_data_df
```

```python
clean_ev_data_df.to_csv('DATA/Clean_Electric_Vehicle_Population_Data.csv', index=False)
```
